```python
import matplotlib.pyplot as plt

anime = ["kuroinu","taboo charmg m","Harem in lyb in aw","High school DxD","Sis X Sis"]

views = [1000,900,700,500,200]

colours = ["#ff4d6d","#c77dff","#4cc9f0","#80ed99","#ffd60a"]

plt.style.use("dark_background")

plt.figure(figsize=(9,16))

bars = plt.bar(anime,views,color=colours)

plt.title("H-anime popularity among gooners",fontsize=20,pad=20)

plt.xlabel("H-Anime",fontsize=14)
plt.ylabel("Popularity",fontsize=14)

for bar in bars:
  yval = bar.get_height()
  plt.text(
      bar.get_x()+bar.get_width()/2,
           yval + 15,
           yval,
           ha='center',
           fontsize=12
      )

  plt.grid(alpha=0.15)

  plt.tight_layout()

  plt.show

```
